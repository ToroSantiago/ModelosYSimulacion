```julia
using Random, Distributions, Plots, StatsPlots
```

```julia
Random.seed!(123)
```

```julia
# Generar 100 datos aleatorios con distribución normal (media=10, desviación=2)
distribucion = Normal(10, 2)
datos = rand(distribucion, 100)
```

```julia
media = mean(datos)
println("A) Media: ", media)
```

```julia
desviacion = std(datos)
println("B) Desvío estándar: ", desviacion)
```

```julia
varianza = var(datos)
println("C) Varianza: ", varianza)
```

```julia
p1 = histogram(datos,bins=15, 
    title="Histograma de datos", 
    xlabel="Valores", 
    ylabel="Frecuencia",
    legend=false, 
    color=:blue,
    alpha=0.7)
```

```julia
p3 = boxplot(["Datos"], datos, 
    title="Diagrama de caja", 
    ylabel="Valores",
    legend=false, 
    color=:green)
```
