```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(123)
```

```python
#100 datos aleatorios con distribución normal (media=10, desviación=2)
datos = np.random.normal(10, 2, 100)
```

```python
#A - CALCULAR LA MEDIA
media = np.mean(datos)
print(f"A) Media: {media}")
```

```python
#B - CALCULAR DESVIO ESTANDAR
desviacion = np.std(datos, ddof=1)  # ddof=1 para estimación insesgada
print(f"B) Desvío estándar: {desviacion}")
```

```python
#C - CALCULAR DESVIO ESTANDAR
varianza = np.var(datos, ddof=1)
print(f"C) Varianza: {varianza}")
```

```python
# Histograma de datos
plt.figure(figsize=(10, 8))

plt.hist(datos, bins=15, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histograma de datos')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')

plt.show()
```

```python
# Grafico de caja
plt.figure(figsize=(10, 8))

# Crear el boxplot
plt.boxplot(datos, patch_artist=True, boxprops=dict(facecolor='green', color='black'))

plt.title('Diagrama de caja')
plt.ylabel('Valores')
plt.xticks([1], ['Datos'])  # Etiquetar el eje x

plt.show()
```
