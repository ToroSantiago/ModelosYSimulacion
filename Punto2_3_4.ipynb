```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
np.random.seed(42)
```

```python
# Tamaño de la muestra
n = 1000
```

```python
# 1. Distribución Uniforme (Min: 0, Max: 1)
datos_uniforme = np.random.uniform(0, 1, n)
```

```python
# 2. Distribución Normal (Media: 0, Desvío: 1)
datos_normal = np.random.normal(0, 1, n)
```

```python
# 3. Distribución Poisson (λ = 6)
datos_poisson = np.random.poisson(lam=6, size=n)
```

```python
# 4. Distribución Exponencial (λ = 34)
datos_exponencial = np.random.exponential(scale=1/34, size=n)
# Nota: En NumPy/SciPy, el parámetro para exponencial es scale=1/λ
```

```python
def calcular_estadisticas(datos, nombre):
    print(f"\nEstadísticas para distribución {nombre}:")
    print(f"Media: {np.mean(datos):.4f}")
    print(f"Desvío estándar: {np.std(datos, ddof=1):.4f}")
    print(f"Varianza: {np.var(datos, ddof=1):.4f}")
    print(f"Mínimo: {np.min(datos):.4f}")
    print(f"Máximo: {np.max(datos):.4f}")
```

```python
calcular_estadisticas(datos_uniforme, "Uniforme")
```

```python
calcular_estadisticas(datos_normal, "Normal")
```

```python
calcular_estadisticas(datos_poisson, "Poisson")
```

```python
calcular_estadisticas(datos_exponencial, "Exponencial")
```

```python
# Crear histogramas para cada distribución
plt.figure(figsize=(12, 10))
```

```python
# 1. Histograma para distribución Uniforme
plt.subplot(2, 2, 1)
plt.hist(datos_uniforme, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Distribución Uniforme (Min: 0, Max: 1)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
```

```python
# 2. Histograma para distribución Normal
plt.subplot(2, 2, 2)
plt.hist(datos_normal, bins=30, alpha=0.7, color='green', edgecolor='black')
plt.title('Distribución Normal (Media: 0, Desvío: 1)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
```

```python
# 3. Histograma para distribución Poisson
plt.subplot(2, 2, 3)
plt.hist(datos_poisson, bins=range(0, max(datos_poisson.astype(int))+2), 
         alpha=0.7, color='red', edgecolor='black')
plt.title('Distribución Poisson (λ = 6)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
```

```python
# 4. Histograma para distribución Exponencial
plt.subplot(2, 2, 4)
plt.hist(datos_exponencial, bins=30, alpha=0.7, color='purple', edgecolor='black')
plt.title('Distribución Exponencial (λ = 34)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
```

```python
# Transformar a distribución Exponencial con λ = 12
lambda_param = 12
datos_exponencial_transformada = -np.log(1 - datos_uniforme) / lambda_param
```

```python
print("Estadísticas de la distribución exponencial transformada (λ = 12):")
print(f"Media teórica: {1/lambda_param:.4f}")
print(f"Media obtenida: {np.mean(datos_exponencial_transformada):.4f}")
print(f"Desvío estándar teórico: {1/lambda_param:.4f}")
print(f"Desvío estándar obtenido: {np.std(datos_exponencial_transformada, ddof=1):.4f}")
```

```python
# 3.b) Gráfico para la variable transformada
plt.figure(figsize=(10, 6))
plt.hist(datos_exponencial_transformada, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)
```

```python
# 3.c) Generar tres muestras con 100, 1000 y 10000 valores (Exponencial con λ = 3)
tamaños = [100, 1000, 10000]
lambda_nuevo = 3
muestras = {}
```

```python
for tamaño in tamaños:
    muestras[tamaño] = np.random.exponential(lambda_nuevo, tamaño)
```

```python
# 3.d) Graficar las tres muestras y observar diferencias
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
```

```python
for tamaño in tamaños:
    plt.figure(figsize=(8, 5))
    plt.hist(muestras[tamaño], bins=30, alpha=0.7, color='green', edgecolor='black', density=True)

    plt.title(f'Muestra Exponencial(λ=3) con n = {tamaño}')
    plt.xlabel('Valores')
    plt.ylabel('Densidad')
    plt.grid(alpha=0.3)
    
    plt.show()
```

```python
for tamaño in tamaños:
    #media muestral
    media = np.mean(muestras[tamaño])
    
    #desvío estándar
    desviacion = np.std(muestras[tamaño], ddof=1)  # ddof=1 para estimación insesgada
    
    #varianza muestral
    varianza = np.var(muestras[tamaño], ddof=1)
    
    print(f"\nEstadísticas para muestra de tamaño {tamaño}:")
    print(f"1. Media muestral: {media:.6f}")
    print(f"2. Desvío estándar muestral: {desviacion:.6f}")
    print(f"3. Varianza muestral: {varianza:.6f}")
```
